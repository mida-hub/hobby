```python
import numpy as np
from scipy.stats import gamma
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

$Gamma(\theta | k, \lambda)\propto \theta^{k-e} \exp(-\lambda \theta)$

```python
eps = 1e-2
lam = 1
k = 11
NMCS = 10000
```

$H = \frac{1}{2} p^2 - (k-1)\log\theta + \lambda \theta$

$\frac {d \theta} {d t} = p$

$\frac {dp}{dt} = (k-1)\frac{1}{\theta} - \lambda$

```python
f = lambda x, lam, k : (k-1) / x - lam
```

```python
x, y = np.linspace(0, 20, 100), np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
CS = plt.contour(X, Y, 0.5 * Y ** 2 - (k - 1) * np.log(X) + lam * X, levels = [-12, -10, -5, 0])
plt.clabel(CS, inline=1, fontsize=10)
```

```python
def eular(q, p, eps, lam, k, f):
    q_new = q + eps * p
    p_new = p + eps * f(q, lam, k)
    q, p = q_new, p_new
    return q, p
```

```python
q, p = 4.0, 0.0
eular_arr = np.zeros([NMCS, 2])
eular_arr[0, 0] = q
eular_arr[0, 1] = p
for i in range(1, NMCS):
    q, p = eular(q, p, eps, lam, k, f)
    eular_arr[i, 0] , eular_arr[i, 1] = q, p

df_euler = pd.DataFrame(eular_arr, columns=["q", "p"])
```

```python
plt.plot(df_euler["q"], df_euler["p"])
CS = plt.contour(X, Y, 0.5 * Y ** 2 - (k - 1) * np.log(X) + lam * X, levels = [-12, -10, -5, 0])
plt.clabel(CS, inline=1, fontsize=10)
```

```python
def leap_flog(q, p, eps, lam, k, f):
    p_new = p + 0.5 * eps * f(q, lam, k)
    q_new = q + eps * p_new
    p_new = p_new + 0.5 * eps * f(q_new, lam, k)
    q, p = q_new, p_new
    return q, p
```

```python
q, p = 4.0, 0.0
lf_arr = np.zeros([NMCS, 2])
lf_arr[0, 0] = q
lf_arr[0, 1] = p
for i in range(1, NMCS):
    q, p = leap_flog(q, p, eps, lam, k, f)
    lf_arr[i, 0] , lf_arr[i, 1] = q, p

df_lf = pd.DataFrame(lf_arr, columns=["q", "p"])
```

```python
plt.plot(df_lf["q"], df_lf["p"])
CS = plt.contour(X, Y, 0.5 * Y ** 2 - (k - 1) * np.log(X) + lam * X, levels = [-12, -10, -5, 0])
plt.clabel(CS, inline=1, fontsize=10)
```
