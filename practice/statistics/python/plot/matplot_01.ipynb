```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'g^')
```

```python
plt.plot(x, y, '--')
```

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, 'bo')
```

```python
plt.plot(x, y)
plt.xlabel('Effort')
plt.ylabel('Earning')
plt.title('This is how your effor earns')
```

```python
plt.plot(x, y, label='Earning with effort')
plt.legend()
plt.xticks(np.arange(-3, 4, 0.5))
plt.yticks([0, 5, 10, 20])
plt.show()
```

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x) * 2
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.legend()
```
